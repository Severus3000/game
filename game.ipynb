```python
import random

# Basic setup
options = ("rock", "paper", "scissors")
total_games = 0
player_wins = 0
computer_wins = 0
ties = 0

# Track choices
player_choices = {"rock": 0, "paper": 0, "scissors": 0}
computer_choices = {"rock": 0, "paper": 0, "scissors": 0}

#prediction function
def prediction(player_choices):
    total_choices = sum(player_choices.values())

    if total_choices == 0:
        return random.choice(options)

    probabilities = {}
    for move, count in player_choices.items():
        probabilities[move] = count / total_choices

    predicted_move = max(probabilities, key=probabilities.get)
    return predicted_move

def counter_move(predicted_move):
    if predicted_move == "rock":
        return "paper"
    elif predicted_move == "paper":
        return "scissors"
    elif predicted_move == "scissors":
        return "rock"

while True:
    # Predict and calculate the computer's move
    predicted_player_choice = prediction(player_choices)
    computer = counter_move(predicted_player_choice)

    # Game input
    player = input("your choice")

 

    # Track choices
    total_games += 1
    player_choices[player] += 1
    computer_choices[computer] += 1

    # gmae output
    if player == computer:
        print(f"Both chose {player}. It's a tie!")
        ties += 1
    elif player == "rock" and computer == "scissors":
        print("player wins")
        player_wins += 1
    elif player == "paper" and computer == "rock":
        print("player wins")
        player_wins += 1
    elif player == "scissors" and computer == "paper":
        print("player wins")
        player_wins += 1
             
    else:
        print("computer wins!")
        computer_wins += 1

    #win rates
    player_win_rate = (player_wins / total_games) * 100
    computer_win_rate = (computer_wins / total_games) * 100
    tie_rate = (ties / total_games) * 100

    # stats
    print("\n--- Game Stats ---")
    print(f"Total games played: {total_games}")
    print(f"Player wins: {player_wins} ({player_win_rate:.2f}%)")
    print(f"Computer wins: {computer_wins} ({computer_win_rate:.2f}%)")
    print(f"Ties: {ties} ({tie_rate:.2f}%)")
    print(f"Player choices: {player_choices}")
    print(f"Computer choices: {computer_choices}")
    print("-------------------\n")
```
